```python
X = train['d_centre'].values
X = X.reshape(len(X),1)
y = train['Price'].values
X_test = df_test['d_centre'].values
X_test = X_test.reshape(len(X_test),1)

from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.19052, random_state=1)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn import linear_model
from sklearn.kernel_ridge import KernelRidge
from sklearn.metrics import mean_squared_error as mse



x_range = np.linspace(X.min(),X.max(),1000).reshape(1000,1)
ridge_regr = linear_model.Ridge(alpha=.5)

####################


clf = KernelRidge(alpha=0.5,kernel = 'rbf')

##############

clf.fit(X_train,y_train)
ridge_regr.fit(X_train,y_train)
y_line_rid = ridge_regr.predict(x_range)
y_line_krid = clf.predict(x_range)

plt.scatter(X_train,y_train,c='g',alpha=0.5)
plt.scatter(X_valid,y_valid,c='y',alpha=0.5)
plt.plot(x_range,y_line_rid,c='r')
plt.plot(x_range,y_line_krid,c='b')


```

```python
y_t = np.linspace(y_train.min(),y_train.max(),1000)
y_v = np.linspace(y_valid.min(),y_valid.max(),1000)

error_rid_train = mse(y_t,y_line_rid)
error_rid_test  = mse(y_v,y_line_rid)

error_krid_train = mse(y_t,y_line_krid)
error_krid_test =mse(y_v,y_line_krid)
print("errores de Ridge ",error_rid_train,' ',error_rid_train)
print("errores de  Kernel Ridge ",error_krid_train,' ',error_krid_test)
```

**PREGUNTA**: ¿Cuál es un mejor modelo para estos datos y por qué?

Es posible observa como el modelo de Ridge se ajusta de una manera mucho mejor a los datos que Kernel ridge, graficando una recta definida. por lo cuál es mejor Ridge.

```python
from sklearn import decomposition
from sklearn.preprocessing import StandardScaler

def do_PCA(n, table, categorical, features=None, show_variance=False):
    pca = decomposition.PCA(n_components=n)
    if features is None:
        scaler = StandardScaler().fit(table)
        stdtab = pd.DataFrame(scaler.transform(table))
    else:
        scaler = StandardScaler().fit(table[features])
        stdtab = pd.DataFrame(scaler.transform(table[features]))
    pca.fit(stdtab)
    table_pca_df = pca.transform(stdtab)
    
    if (show_variance):
        varx = pca.explained_variance_ratio_
        plt.plot(np.arange(1,n+1),varx.cumsum())
        plt.ylabel("% of variance")
        plt.xlabel("components")

    ind_pca = []
    for i in range(n):
        ind_pca.append("PC"+str(i+1))
    
    table_pca = pd.DataFrame(table_pca_df, columns=ind_pca).dropna()
    PCA_features = [categorical]
    for i in range(n):
        table["PC"+str(i+1)] = table_pca["PC"+str(i+1)]
        PCA_features.append("PC"+str(i+1))
        
do_PCA(4,train,'Price',top_feats,show_variance=True)

```
